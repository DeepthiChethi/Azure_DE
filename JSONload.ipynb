```python
from pyspark.sql.functions import explode, col
```

```python
json_path = "Files/Sample.JSON"
```

```python
df = spark.read.json(json_path, multiLine=True)
df.show(truncate=False)
```

```python
df_exploded = df.withColumn("item", explode(col("items")))
df_flat = df_exploded.select(
    col("id"),
    col("name"),
    col("item.id").alias("item_id"),
    col("item.name").alias("item_name"),
    col("item.price").alias("item_price")
)

df_flat.show(truncate=False)
```

```python
df_flat.write.mode("overwrite").option("header", "true").csv("Files/samplejson.csv")
```
